```python
from decimal import Decimal


def quadratic(a, b, c):

    # The discriminant is first calculated using the Decimal function that assures safe subtraction as adviced in David  Goldberg's book.
    discriminant = Decimal(b**2) - Decimal(4 * a * c)
    
    # Then numpy is used to provide a check whether the discriminant is less than zero and output 'No Real Roots' if so.
    if np.any(discriminant < 0):
        return "No Real Roots"
    
    # The first root is calculated using the quadratic formula.
    x1 = np.where(b < 0, (Decimal(-b) + np.sqrt(discriminant)) / (2 * Decimal(a)), (Decimal(-b) - np.sqrt(discriminant)) / (2 * Decimal(a)))
    
    # The second root is calculted from the first root.
    x2 = Decimal(c) / (Decimal(a) * x1)
    
    return (x1, x2)


```

```python
import numpy as np
t_max = np.random.randint(0, 101)
random_numbers = np.random.randint(0, 2, size=t_max)
steps = np.where(random_numbers == 0, -1, +1)
values = np.cumsum(steps)
values
```
